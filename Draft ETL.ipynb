```python
country_data_file = "countries of the world.csv"
country_data_file_df = pd.read_csv(country_data_file)
country_data_file_df.head()
```

```python
Cost_of_Living_file = "Cost_of_Living_Index_2022.csv"
Cost_of_Living_df = pd.read_csv(Cost_of_Living_file)
Cost_of_Living_df.head()
```

```python
# Create a filtered dataframe from specific columns
country_data_cols = ["Country", 
                     "Region", 
                     "Population", 
                     "Pop. Density (per sq. mi.)", 
                     "Net migration", 
                     "Infant mortality (per 1000 births)", 
                     "GDP ($ per capita)", 
                     "Literacy (%)", 
                     "Phones (per 1000)", 
                     "Birthrate", 
                     "Deathrate"]

country_data_transformed = country_data_file_df[country_data_cols].copy()

# Rename the column headers
country_data_transformed = country_data_transformed.rename(columns={"Country": "Country",
                                                                    "Region": "Region",
                                                                    "Population": "Population", 
                                                                    "Pop. Density (per sq. mi.)": "Population_Density_(/sqr-mile)",
                                                                    "Net migration": "Net_Migration",
                                                                    "Infant mortality (per 1000 births)": "Infant_Mortality_(per_1000_births)",
                                                                    "GDP ($ per capita)": "GDP_($_per_capita)",
                                                                    "Literacy (%)": "Literacy_(%)",
                                                                    "Phones (per 1000)": "Phones_(per_1000)",
                                                                    "Birthrate": "Birthrate",
                                                                    "Deathrate": "Deathrate"
                                                                   })

# Clean the data by dropping duplicates and setting the index
# country_data_transformed.drop_duplicates("Country", inplace=True)
# country_data_transformed.set_index("Country", inplace=True)

country_data_transformed.head()
```
